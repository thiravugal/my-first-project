# üß© PART 1 --- Procedural Version

-   Store account holder name

## ‚úÖ C++ Procedural Version

``` cpp
#include <iostream>
using namespace std;

string accountHolder = "John";
double balance = 1000;

void deposit(double amount) {
    balance += amount;
}

void withdraw(double amount) {
    if (amount <= balance)
        balance -= amount;
    else
        cout << "Insufficient balance\n";
}

void checkBalance() {
    cout << "Current Balance: " << balance << endl;
}

int main() {
    int choice;
    do {
        cout << "\n1.Deposit 2.Withdraw 3.Check Balance 4.Exit\n";
        cin >> choice;

        if (choice == 1) {
            double amt;
            cout << "Enter amount: ";
            cin >> amt;
            deposit(amt);
        }
        else if (choice == 2) {
            double amt;
            cout << "Enter amount: ";
            cin >> amt;
            withdraw(amt);
        }
        else if (choice == 3) {
            checkBalance();
        }

    } while (choice != 4);

    return 0;
}
```

------------------------------------------------------------------------

## ‚úÖ Java Procedural Version

``` java
import java.util.Scanner;

public class BankProcedural {

    static String accountHolder = "John";
    static double balance = 1000;

    static void deposit(double amount) {
        balance += amount;
    }

    static void withdraw(double amount) {
        if (amount <= balance)
            balance -= amount;
        else
            System.out.println("Insufficient balance");
    }

    static void checkBalance() {
        System.out.println("Current Balance: " + balance);
    }

    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        int choice;

        do {
            System.out.println("\n1.Deposit 2.Withdraw 3.Check Balance 4.Exit");
            choice = sc.nextInt();

            if (choice == 1) {
                System.out.print("Enter amount: ");
                deposit(sc.nextDouble());
            }
            else if (choice == 2) {
                System.out.print("Enter amount: ");
                withdraw(sc.nextDouble());
            }
            else if (choice == 3) {
                checkBalance();
            }

        } while (choice != 4);

        sc.close();
    }
}
```

# üèó PART 2 --- Convert to OOP

## Requirements

-   Create class `BankAccount`
-   Make balance private (Encapsulation)
-   Create deposit(), withdraw(), checkBalance()
-   Create multiple objects

------------------------------------------------------------------------

## ‚úÖ C++ OOP Version

``` cpp
#include <iostream>
using namespace std;

class BankAccount {
private:
    string accountHolder;
    double balance;

public:
    BankAccount(string name, double bal) {
        accountHolder = name;
        balance = bal;
    }

    void deposit(double amount) {
        balance += amount;
    }

    void withdraw(double amount) {
        if (amount <= balance)
            balance -= amount;
        else
            cout << "Insufficient balance\n";
    }

    void checkBalance() {
        cout << accountHolder << "'s Balance: " << balance << endl;
    }
};

int main() {
    BankAccount acc1("John", 1000);
    BankAccount acc2("Alice", 500);

    acc1.deposit(200);
    acc1.checkBalance();

    acc2.withdraw(100);
    acc2.checkBalance();

    return 0;
}
```

------------------------------------------------------------------------

## ‚úÖ Java OOP Version

``` java
class BankAccount {
    private String accountHolder;
    private double balance;

    public BankAccount(String name, double bal) {
        accountHolder = name;
        balance = bal;
    }

    public void deposit(double amount) {
        balance += amount;
    }

    public void withdraw(double amount) {
        if (amount <= balance)
            balance -= amount;
        else
            System.out.println("Insufficient balance");
    }

    public void checkBalance() {
        System.out.println(accountHolder + "'s Balance: " + balance);
    }
}

public class BankOOP {
    public static void main(String[] args) {
        BankAccount acc1 = new BankAccount("John", 1000);
        BankAccount acc2 = new BankAccount("Alice", 500);

        acc1.deposit(200);
        acc1.checkBalance();

        acc2.withdraw(100);
        acc2.checkBalance();
    }
}
```

# üß¨ PART 3 --- Inheritance (Advanced)

## Add SavingsAccount

### C++

``` cpp
class SavingsAccount : public BankAccount {
private:
    double interestRate;

public:
    SavingsAccount(string name, double bal, double rate)
        : BankAccount(name, bal) {
        interestRate = rate;
    }
};
```

### Java

``` java
class SavingsAccount extends BankAccount {
    private double interestRate;

    public SavingsAccount(String name, double bal, double rate) {
        super(name, bal);
        interestRate = rate;
    }
}
```
